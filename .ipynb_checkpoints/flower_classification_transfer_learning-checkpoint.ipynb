## Part A — Dataset Handling from Disk

1. Create balanced subsets per class.
2. Count and print images per class.
3. Display one random image per class.
4. Implement `make_subset(src_dir, dst_dir, per_class=100, seed=42)`.
5. Print total number of images and classes inside subset folder.


```python

# Part A: imports and helper functions
import os
import random
import shutil
import numpy as np
from matplotlib import pyplot as plt
from pathlib import Path
from collections import Counter
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline

# Set paths (edit these to match where you extracted the dataset)
# Example: DATA_DIR = "C:/datasets/flowers" or "./flowers"
DATA_DIR = "flowers"  # <-- change if your dataset is elsewhere
SUBSET_BASE = "subsets"  # destination folder for subsets

def count_images_per_class(src_dir):
    """Traverse src_dir and count images in each immediate subfolder (class)."""
    src = Path(src_dir)
    classes = [p for p in src.iterdir() if p.is_dir()]
    counts = {}
    for c in classes:
        counts[c.name] = len(list(c.glob('*.*')))
    return counts

def show_random_images(src_dir, classes=None, figsize=(12,6)):
    """Display one random image from each class using matplotlib."""
    src = Path(src_dir)
    classes = classes or [p.name for p in src.iterdir() if p.is_dir()]
    n = len(classes)
    cols = min(5, n)
    rows = (n + cols - 1) // cols
    plt.figure(figsize=figsize)
    i = 1
    for cls in classes:
        cls_dir = src/cls
        imgs = list(cls_dir.glob('*.*'))
        if not imgs:
            continue
        img_path = random.choice(imgs)
        img = Image.open(img_path).convert('RGB')
        plt.subplot(rows, cols, i)
        plt.imshow(img)
        plt.title(cls)
        plt.axis('off')
        i += 1
    plt.tight_layout()
    plt.show()

def make_subset(src_dir, dst_dir, per_class=100, seed=42, overwrite=False):
    """Create balanced subset by copying per_class images per class from src_dir to dst_dir.""" 
    random.seed(seed)
    src = Path(src_dir)
    dst = Path(dst_dir)
    if overwrite and dst.exists():
        shutil.rmtree(dst)
    dst.mkdir(parents=True, exist_ok=True)
    classes = [p for p in src.iterdir() if p.is_dir()]
    for c in classes:
        src_files = [p for p in c.glob('*.*')]
        if len(src_files) == 0:
            print(f"Warning: class {c.name} has no images.")
            continue
        chosen = random.sample(src_files, min(per_class, len(src_files)))
        class_dst = dst / c.name
        class_dst.mkdir(parents=True, exist_ok=True)
        for f in chosen:
            shutil.copy(f, class_dst / f.name)
    # return summary
    total = sum(len(list((dst / c.name).glob('*.*'))) for c in classes)
    return {'classes': [c.name for c in classes], 'per_class_requested': per_class, 'total_copied': total}
```

```python

# Example usage for Part A.
# Make sure DATA_DIR points to your extracted 'flowers' directory.
print('Counts in original dataset:')
print(count_images_per_class(DATA_DIR))

# Display one random image per class
show_random_images(DATA_DIR)

# Create a subset with 200 images per class
res = make_subset(DATA_DIR, os.path.join(SUBSET_BASE, 'subset_200'), per_class=200, seed=42, overwrite=True)
print('\nSubset creation summary:')
print(res)

# Verify counts in the subset folder
print('\nCounts in subset_200:')
print(count_images_per_class(os.path.join(SUBSET_BASE, 'subset_200')))
```

## Part B — Data Loading and Augmentation

6. Load the subset from disk using `tf.keras.utils.image_dataset_from_directory()` and split using `validation_split`.
7. Create a data augmentation pipeline using `tf.keras.Sequential()` with `RandomFlip`, `RandomRotation`, and `RandomZoom`.
Visualize 5 augmented versions of a single flower image.

```python

# Part B: Data Loading & Augmentation
import tensorflow as tf

AUTOTUNE = tf.data.AUTOTUNE
IMG_SIZE = (224, 224)  # MobileNetV2 default input size

def load_datasets_from_directory(subset_dir, batch_size=32, img_size=IMG_SIZE, seed=42, validation_split=0.2):
    """Load train/val datasets from disk using validation_split of image_dataset_from_directory."""
    train_ds = tf.keras.utils.image_dataset_from_directory(
        subset_dir,
        validation_split=validation_split,
        subset='training',
        seed=seed,
        image_size=img_size,
        batch_size=batch_size
    )
    val_ds = tf.keras.utils.image_dataset_from_directory(
        subset_dir,
        validation_split=validation_split,
        subset='validation',
        seed=seed,
        image_size=img_size,
        batch_size=batch_size
    )
    class_names = train_ds.class_names
    # Prefetch for performance
    train_ds = train_ds.prefetch(buffer_size=AUTOTUNE)
    val_ds = val_ds.prefetch(buffer_size=AUTOTUNE)
    return train_ds, val_ds, class_names

# Augmentation pipeline
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip('horizontal'),
    tf.keras.layers.RandomRotation(0.15),
    tf.keras.layers.RandomZoom(0.15)
], name='data_augmentation')

# Visualize 5 augmented versions of a single image
import numpy as np
from matplotlib import pyplot as plt

def visualize_augmentations(image_path, augmentation_model, num=5, img_size=IMG_SIZE):
    img = tf.keras.preprocessing.image.load_img(image_path, target_size=img_size)
    img_arr = tf.keras.preprocessing.image.img_to_array(img)
    img_batch = tf.expand_dims(img_arr, 0)  # create batch
    plt.figure(figsize=(12,3))
    for i in range(num):
        augmented = augmentation_model(img_batch, training=True)
        ax = plt.subplot(1, num, i+1)
        plt.imshow(tf.cast(augmented[0] / 255.0, tf.float32))
        plt.axis('off')
    plt.suptitle('Augmented versions')
    plt.show()

# Usage note:
print('To load subset, call load_datasets_from_directory("subsets/subset_200") after creating it.')
```

## Part C — Model Development (From Scratch)

8. Build a simple CNN with 3 conv blocks, Flatten, Dense(128, ReLU), and Dense(5, softmax).
9. Compile with adam + categorical_crossentropy.
10. Train for 20–40 epochs and plot accuracy vs validation accuracy.

```python

# Part C: Build a simple CNN from scratch
from tensorflow.keras import layers, models

def make_scratch_model(input_shape=IMG_SIZE + (3,), num_classes=5):
    inputs = layers.Input(shape=input_shape)
    x = layers.Rescaling(1./255)(inputs)
    # Block 1
    x = layers.Conv2D(32, 3, padding='same', activation='relu')(x)
    x = layers.MaxPooling2D()(x)
    # Block 2
    x = layers.Conv2D(64, 3, padding='same', activation='relu')(x)
    x = layers.MaxPooling2D()(x)
    # Block 3
    x = layers.Conv2D(128, 3, padding='same', activation='relu')(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Flatten()(x)
    x = layers.Dense(128, activation='relu')(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    model = models.Model(inputs, outputs, name='scratch_cnn')
    return model

# Compile
model = make_scratch_model()
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Training template (uncomment and run after loading datasets):
# epochs = 20
# history = model.fit(train_ds, validation_data=val_ds, epochs=epochs)
# Plotting training history:
# import pandas as pd
# def plot_history(history):
#     import matplotlib.pyplot as plt
#     plt.figure(figsize=(8,4))
#     plt.plot(history.history['accuracy'], label='train_acc')
#     plt.plot(history.history['val_accuracy'], label='val_acc')
#     plt.xlabel('Epoch')
#     plt.ylabel('Accuracy')
#     plt.legend()
#     plt.show()
# plot_history(history)
```

## Part D — Transfer Learning

11. Use MobileNetV2 (or ResNet50) as a frozen base (weights='imagenet', include_top=False).
12. Fine-tune the top 20 layers and compare accuracy.
13. Repeat experiments with different subset sizes.

```python

# Part D: Transfer learning templates
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models

def make_transfer_model(img_size=IMG_SIZE, num_classes=5, base_trainable=False):
    base = MobileNetV2(weights='imagenet', include_top=False, input_shape=img_size + (3,))
    base.trainable = base_trainable
    inputs = layers.Input(shape=img_size + (3,))
    x = layers.Rescaling(1./255)(inputs)
    x = base(x, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    model = models.Model(inputs, outputs, name='mobilenetv2_transfer')
    return model

# Frozen model (train only head)
frozen_model = make_transfer_model(base_trainable=False)
frozen_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
frozen_model.summary()

# Fine-tune top 20 layers (unfreeze)
def fine_tune_model(model, base_layers_to_unfreeze=20):
    # attempt to locate MobileNetV2 base in model
    base = None
    for layer in model.layers:
        if hasattr(layer, 'output') and 'MobilenetV2' in layer.name.lower() or isinstance(layer, tf.keras.applications.MobileNetV2):
            base = layer
            break
    # fallback: assume layer index 2 is base (may vary)
    if base is None:
        base = model.layers[2]
    # set trainability
    total_layers = len(base.layers)
    for layer in base.layers[:-base_layers_to_unfreeze]:
        layer.trainable = False
    for layer in base.layers[-base_layers_to_unfreeze:]:
        layer.trainable = True
    # recompile with lower LR
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-5),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

print('Transfer learning templates ready. To train, create datasets with load_datasets_from_directory() and call model.fit()')
```

## Part E — Evaluation & Reporting

14. Record training accuracy, validation accuracy, and model size (MB).
15. Plot validation accuracy vs subset size for Scratch, Transfer (frozen), Transfer (fine-tuned).
16. Generate confusion matrix and classification report.
17. Save best model as `best_flower_model.keras` and provide prediction function.

```python

# Part E: Evaluation & Reporting templates
import numpy as np
import os
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf

def model_size_mb(model_path):
    """Return file size in MB for saved model directory/file."""
    if os.path.isdir(model_path):
        # size of directory
        total = 0
        for dirpath, dirnames, filenames in os.walk(model_path):
            for f in filenames:
                fp = os.path.join(dirpath, f)
                total += os.path.getsize(fp)
        return total / (1024*1024)
    else:
        return os.path.getsize(model_path) / (1024*1024)

def evaluate_and_report(model, val_ds, class_names):
    # Collect true labels and predictions
    y_true = []
    y_pred = []
    for images, labels in val_ds:
        preds = model.predict(images)
        preds_labels = np.argmax(preds, axis=1)
        y_true.extend(labels.numpy().tolist())
        y_pred.extend(preds_labels.tolist())
    print('Classification Report:')
    print(classification_report(y_true, y_pred, target_names=class_names))
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()

def save_best_model(model, path='best_flower_model.keras'):
    model.save(path)
    print('Saved model to', path)
    print('Model size (MB):', model_size_mb(path))

def predict_image(model, image_path, class_names, img_size=IMG_SIZE):
    img = tf.keras.preprocessing.image.load_img(image_path, target_size=img_size)
    arr = tf.keras.preprocessing.image.img_to_array(img)
    arr = arr / 255.0
    arr = np.expand_dims(arr, 0)
    preds = model.predict(arr)
    idx = np.argmax(preds[0])
    return class_names[idx], preds[0][idx]

# Usage example (run after training and when you have val_ds and class_names):
# evaluate_and_report(best_model, val_ds, class_names)
# save_best_model(best_model, 'best_flower_model.keras')
# print(predict_image(best_model, 'path/to/some/image.jpg', class_names))
```

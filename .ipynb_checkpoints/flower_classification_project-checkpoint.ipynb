## Part A — Dataset Handling from Disk

This cell:
1. Counts images per class.
2. Shows one random image per class.
3. Implements `make_subset(src_dir, dst_dir, per_class=100, seed=42)` to create balanced subsets.
4. Prints total images/classes inside the subset.


```python
import os, random, shutil
from pathlib import Path
import matplotlib.pyplot as plt
from PIL import Image

# USER: set this to the path where you extracted flower_photos
DATA_DIR = Path("flowers")  # <-- change if needed, e.g., Path("/kaggle/input/flower_photos/flowers")

# 1) Count and print number of images per class by traversing folders
def count_images_per_class(data_dir):
    classes = []
    counts = {}
    for sub in sorted([p for p in data_dir.iterdir() if p.is_dir()]):
        name = sub.name
        classes.append(name)
        cnt = len(list(sub.glob("*.jpg"))) + len(list(sub.glob("*.jpeg"))) + len(list(sub.glob("*.png")))
        counts[name] = cnt
    return classes, counts

classes, counts = count_images_per_class(DATA_DIR)
print("Found classes and counts:")
for k,v in counts.items():
    print(f" - {k}: {v} images")
print()

# 2) Display one random image from each class
plt.figure(figsize=(12,6))
for i,c in enumerate(classes):
    imgs = list((DATA_DIR / c).glob("*.*"))
    if not imgs:
        continue
    p = random.choice(imgs)
    img = Image.open(p).convert("RGB")
    plt.subplot(1, len(classes), i+1)
    plt.imshow(img)
    plt.axis('off')
    plt.title(c)
plt.show()

# 4) make_subset function
def make_subset(src_dir, dst_dir, per_class=100, seed=42, overwrite=False):
    random.seed(seed)
    src_dir = Path(src_dir)
    dst_dir = Path(dst_dir)
    if dst_dir.exists() and overwrite:
        shutil.rmtree(dst_dir)
    dst_dir.mkdir(parents=True, exist_ok=True)
    classes = [p for p in sorted(src_dir.iterdir()) if p.is_dir()]
    for c in classes:
        src_class_dir = src_dir / c.name
        dst_class_dir = dst_dir / c.name
        dst_class_dir.mkdir(parents=True, exist_ok=True)
        imgs = [p for p in src_class_dir.glob("*.*") if p.suffix.lower() in [".jpg",".jpeg",".png"]]
        if len(imgs) < per_class:
            print(f"Warning: class {c.name} has only {len(imgs)} images (<{per_class})")
        chosen = random.sample(imgs, min(per_class, len(imgs)))
        for p in chosen:
            shutil.copy(p, dst_class_dir / p.name)
    # Print verification
    total = sum(len(list((dst_dir / c.name).glob("*.*"))) for c in classes)
    print(f"Created subset at {dst_dir} with {len(classes)} classes and {total} images")
    return dst_dir

# Example: create subset_200 with 200 images per class
subset200 = make_subset(DATA_DIR, Path("subset_200"), per_class=200, seed=42)
```

## Part B — Data Loading and Augmentation

This cell loads the subset using `tf.keras.preprocessing.image_dataset_from_directory`
(or `tf.keras.utils.image_dataset_from_directory` depending on TF version) and builds
a data augmentation pipeline.


```python
import tensorflow as tf
from tensorflow.keras import layers

# Parameters
BATCH_SIZE = 32
IMG_SIZE = (160,160)  # good for MobileNetV2
AUTOTUNE = tf.data.AUTOTUNE
SUBSET_DIR = Path("subset_200")  # change as needed (e.g., subset_100, subset_50)

# 6) Load with train/validation split
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    SUBSET_DIR,
    validation_split=0.2,
    subset="training",
    seed=42,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    SUBSET_DIR,
    validation_split=0.2,
    subset="validation",
    seed=42,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
)

class_names = train_ds.class_names
print("Classes:", class_names)

# Prefetch for performance
train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)

# 7) Data augmentation pipeline
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.15),
    layers.RandomZoom(0.1),
], name="data_augmentation")

# Visualize 5 augmented versions of a single example
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
for images, labels in train_ds.take(1):
    img = images[0]
    for i in range(5):
        augmented = data_augmentation(tf.expand_dims(img,0), training=True)
        ax = plt.subplot(1,5,i+1)
        plt.imshow(tf.cast(augmented[0], tf.uint8).numpy())
        plt.axis('off')
plt.show()
```

## Part C — Model Development (Scratch CNN)

```python
from tensorflow.keras import layers, models, optimizers

# 8) Build a simple CNN
def build_scratch_cnn(input_shape=IMG_SIZE+(3,), num_classes=5):
    inputs = layers.Input(shape=input_shape)
    x = layers.Rescaling(1./255)(inputs)
    x = layers.Conv2D(32, 3, activation='relu', padding='same')(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Conv2D(64, 3, activation='relu', padding='same')(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Conv2D(128, 3, activation='relu', padding='same')(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Flatten()(x)
    x = layers.Dense(128, activation='relu')(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    model = models.Model(inputs, outputs, name="scratch_cnn")
    return model

scratch_model = build_scratch_cnn()
scratch_model.summary()

# 9) Compile
scratch_model.compile(optimizer='adam',
                      loss='sparse_categorical_crossentropy',
                      metrics=['accuracy'])

# 10) Train for 20 epochs (change as desired)
EPOCHS = 20
history_scratch = scratch_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS
)

# Plot training vs validation accuracy
import matplotlib.pyplot as plt
acc = history_scratch.history['accuracy']
val_acc = history_scratch.history['val_accuracy']
epochs_range = range(1, len(acc)+1)
plt.figure(figsize=(8,4))
plt.plot(epochs_range, acc, label='Train Acc')
plt.plot(epochs_range, val_acc, label='Val Acc')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Scratch CNN Accuracy')
plt.show()
```

## Part D — Transfer Learning (MobileNetV2)

1. Use MobileNetV2 as a frozen base.
2. Train new classification head.
3. Fine-tune top 20 layers and compare.


```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import Model

IMG_SHAPE = IMG_SIZE + (3,)
base_model = MobileNetV2(input_shape=IMG_SHAPE, include_top=False, weights='imagenet')
base_model.trainable = False  # frozen

# Build head
inputs = layers.Input(shape=IMG_SHAPE)
x = data_augmentation(inputs)  # include augmentation in the model
# MobileNetV2 expects inputs in [-1,1]
x = layers.Rescaling(1./127.5)(x)
x = layers.Rescaling(0.5, offset=-1)(x)
x = base_model(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.2)(x)
outputs = layers.Dense(len(class_names), activation='softmax')(x)
model_tl = Model(inputs, outputs)

model_tl.compile(optimizer='adam',
                 loss='sparse_categorical_crossentropy',
                 metrics=['accuracy'])

model_tl.summary()

# Train the frozen model head
EPOCHS = 20
history_frozen = model_tl.fit(train_ds, validation_data=val_ds, epochs=EPOCHS)

# Fine-tune: unfreeze top 20 layers of base_model
base_model.trainable = True
# Freeze all layers except the top 20
for layer in base_model.layers[:-20]:
    layer.trainable = False

model_tl.compile(optimizer=tf.keras.optimizers.Adam(1e-5),
                 loss='sparse_categorical_crossentropy',
                 metrics=['accuracy'])

EPOCHS_FINE = 20
history_finetune = model_tl.fit(train_ds, validation_data=val_ds, epochs=EPOCHS_FINE)

# Compare validation accuracies
import matplotlib.pyplot as plt
v1 = history_frozen.history['val_accuracy']
v2 = history_finetune.history['val_accuracy']
plt.plot(range(1,len(v1)+1), v1, label='Frozen val_acc')
plt.plot(range(1,len(v2)+1), v2, label='Finetune val_acc')
plt.legend(); plt.xlabel('Epoch'); plt.ylabel('Val Accuracy'); plt.title('Transfer Learning: Frozen vs Fine-tune'); plt.show()
```

## Part E — Evaluation & Reporting

- Record training & validation accuracy and model size.
- Plot Validation Accuracy vs Subset Size for Scratch, Transfer (frozen), Transfer (fine-tuned).
- Confusion matrix & classification report for best model.
- Save best model and write a small prediction script.


```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import tensorflow as tf
import os

# 14) Record final accuracies and model sizes
def get_final_metrics(history, model):
    final_train_acc = history.history.get('accuracy', [None])[-1]
    final_val_acc = history.history.get('val_accuracy', [None])[-1]
    # model size (approx) - save temporarily
    tmp_path = "tmp_save.keras"
    model.save(tmp_path, include_optimizer=False)
    size_mb = os.path.getsize(tmp_path) / (1024*1024)
    try:
        os.remove(tmp_path)
    except:
        pass
    return final_train_acc, final_val_acc, size_mb

metrics_scratch = get_final_metrics(history_scratch, scratch_model)
metrics_frozen = get_final_metrics(history_frozen, model_tl)  # after frozen training (model_tl currently fine-tuned, but it's ok)
metrics_finetune = get_final_metrics(history_finetune, model_tl)

print("Scratch model metrics (train_acc, val_acc, size_mb):", metrics_scratch)
print("Transfer frozen metrics (train_acc, val_acc, size_mb):", metrics_frozen)
print("Transfer finetune metrics (train_acc, val_acc, size_mb):", metrics_finetune)

# 16) Confusion matrix and classification report for the best model (choose model_tl as example)
# Build predictions on validation set
y_true = []
y_pred = []
for images, labels in val_ds:
    preds = model_tl.predict(images)
    y_true.extend(labels.numpy().tolist())
    y_pred.extend(np.argmax(preds, axis=1).tolist())

cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:\n", cm)
print("\nClassification Report:\n", classification_report(y_true, y_pred, target_names=class_names))

# 17) Save best model and small predict function
BEST_PATH = "best_flower_model.keras"
model_tl.save(BEST_PATH)
print(f"Saved best model to {BEST_PATH} (size: {os.path.getsize(BEST_PATH)/(1024*1024):.2f} MB)")

# Small predict function
from PIL import Image
def predict_image(model_path, image_path, img_size=IMG_SIZE):
    model = tf.keras.models.load_model(model_path)
    img = Image.open(image_path).convert("RGB").resize(img_size)
    arr = np.array(img)[None,...] / 255.0
    preds = model.predict(arr)
    idx = np.argmax(preds[0])
    return class_names[idx], preds[0][idx]

# Example usage (uncomment and change path):
# print(predict_image(BEST_PATH, "path_to_new_flower.jpg"))
```
